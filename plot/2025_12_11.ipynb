```python
# tuning with noisy
df = pd.read_csv(
    os.path.join(root, "2025_12_11", "exp_1", "run_0", 
                 "seed=0.csv")
)
tune_obj_arr = df['true value'].to_numpy()
tune_lb_arr = df['true opt_lb'].to_numpy()
tune_gap_arr = tune_obj_arr - tune_lb_arr
tune_noisy_obj_arr = df['point value'].to_numpy()
tune_noisy_lb_arr = df['point opt_lb'].to_numpy()

df = pd.read_csv(
    os.path.join(root, "2025_12_11", "exp_1", "run_0", 
                 "tuning_map_seed=0.csv")
)
tune_id_arr = df['tune_id'].to_numpy()
tune_n_iter_arr = df['n_iter'].to_numpy()
tune_explore_T_arr = df['explore_T'].to_numpy()

# tuning with exact
df = pd.read_csv(
    os.path.join(root, "2025_12_11", "exp_1", "run_1", 
                 "seed=0.csv")
)
alt_tune_obj_arr = df['true value'].to_numpy()
alt_tune_lb_arr = df['true opt_lb'].to_numpy()
alt_tune_gap_arr = alt_tune_obj_arr - alt_tune_lb_arr
alt_tune_noisy_obj_arr = df['point value'].to_numpy()
alt_tune_noisy_lb_arr = df['point opt_lb'].to_numpy()

df = pd.read_csv(
    os.path.join(root, "2025_12_11", "exp_1", "run_1", 
                 "tuning_map_seed=0.csv")
)
alt_tune_id_arr = df['tune_id'].to_numpy()
alt_tune_n_iter_arr = df['n_iter'].to_numpy()
alt_tune_explore_T_arr = df['explore_T'].to_numpy()
```

## Data cleaning for noisy tuning

```python
# tuning with noisy cost
n_tune = len(np.unique(tune_id_arr))
curr_iter_arr = np.zeros(n_tune, dtype=int)
obj_multiarr = np.zeros((n_tune, len(tune_obj_arr)))
lb_multiarr = np.zeros(obj_multiarr.shape)
gap_multiarr = np.zeros(obj_multiarr.shape)
noisy_obj_multiarr = np.zeros(obj_multiarr.shape)
noisy_lb_multiarr = np.zeros(obj_multiarr.shape)
cum_sample_arr = np.zeros(int(1e6))

ct = 0
for i, tune_id in enumerate(tune_id_arr):
    curr_iters = curr_iter_arr[tune_id]
    n_iters = tune_n_iter_arr[i]
    
    obj_multiarr[tune_id,curr_iters:curr_iters+n_iters] = tune_obj_arr[ct:ct+n_iters]
    lb_multiarr[tune_id,curr_iters:curr_iters+n_iters] = tune_lb_arr[ct:ct+n_iters]
    noisy_obj_multiarr[tune_id,curr_iters:curr_iters+n_iters] = tune_noisy_obj_arr[ct:ct+n_iters]
    noisy_lb_multiarr[tune_id,curr_iters:curr_iters+n_iters] = tune_noisy_lb_arr[ct:ct+n_iters]

    curr_iter_arr[tune_id] = curr_iters+n_iters

    cum_sample_arr[1+ct:1+ct+n_iters] = cum_sample_arr[ct] + tune_explore_T_arr[tune_id] * np.arange(1,n_iters+1)
    
    ct += n_iters
    
cum_sample_arr = cum_sample_arr[1:ct+1]
gap_multiarr = obj_multiarr - lb_multiarr
noisy_gap_multiarr = noisy_obj_multiarr - noisy_lb_multiarr

tune_sample_arr = np.zeros(obj_multiarr.shape)
for i in range(n_tune):
    tune_sample_arr[i,0:curr_iter_arr[i]] = tune_explore_T_arr[i] * np.arange(curr_iter_arr[i])

# tuning with exact cost
n_tune = len(np.unique(alt_tune_id_arr))
alt_curr_iter_arr = np.zeros(n_tune, dtype=int)
alt_obj_multiarr = np.zeros((n_tune, len(tune_obj_arr)))
alt_lb_multiarr = np.zeros(obj_multiarr.shape)
alt_gap_multiarr = np.zeros(obj_multiarr.shape)
alt_noisy_obj_multiarr = np.zeros(obj_multiarr.shape)
alt_noisy_lb_multiarr = np.zeros(obj_multiarr.shape)
alt_cum_sample_arr = np.zeros(int(1e6))

ct = 0
for i, tune_id in enumerate(alt_tune_id_arr):
    curr_iters = alt_curr_iter_arr[tune_id]
    n_iters = alt_tune_n_iter_arr[i]
    
    alt_obj_multiarr[tune_id,curr_iters:curr_iters+n_iters] = alt_tune_obj_arr[ct:ct+n_iters]
    alt_lb_multiarr[tune_id,curr_iters:curr_iters+n_iters] = alt_tune_lb_arr[ct:ct+n_iters]
    alt_noisy_obj_multiarr[tune_id,curr_iters:curr_iters+n_iters] = alt_tune_noisy_obj_arr[ct:ct+n_iters]
    alt_noisy_lb_multiarr[tune_id,curr_iters:curr_iters+n_iters] = alt_tune_noisy_lb_arr[ct:ct+n_iters]

    alt_curr_iter_arr[tune_id] = curr_iters+n_iters

    alt_cum_sample_arr[1+ct:1+ct+n_iters] = alt_cum_sample_arr[ct] + alt_tune_explore_T_arr[tune_id] * np.arange(1,n_iters+1)
    
    ct += n_iters

alt_cum_sample_arr = alt_cum_sample_arr[1:ct+1]
alt_gap_multiarr = alt_obj_multiarr - alt_lb_multiarr
alt_noisy_gap_multiarr = alt_noisy_obj_multiarr - alt_noisy_lb_multiarr

alt_tune_sample_arr = np.zeros(obj_multiarr.shape)
for i in range(n_tune):
    alt_tune_sample_arr[i,0:alt_curr_iter_arr[i]] = alt_tune_explore_T_arr[i] * np.arange(alt_curr_iter_arr[i])

print(curr_iter_arr)
print(alt_curr_iter_arr)
```

```python
# true cost
plt.style.use('ggplot')
_, ax = plt.subplots()

for i in range(n_tune):
    curr_iter = curr_iter_arr[i]
    ax.plot(tune_sample_arr[i,:curr_iter], noisy_gap_multiarr[i,:curr_iter], label="tune_id=%d" % i,
           color=color_arr[i], linestyle=lss_arr[i])
    
ax.set(
    title="SPMD noisy tuning performance\n on a simple Battery problem",
    ylabel="Estimated optimality gap",
    xlabel="Samples",
    xlim=(-1e5,4e6),
)
ax.legend()
```

```python
# true cost
i = np.argmax(curr_iter_arr)
print("best performing tune_id = %d with T=%d" % (i, tune_explore_T_arr[i]))

plt.style.use('ggplot')
_, ax = plt.subplots()

for i in range(n_tune):
    curr_iter = curr_iter_arr[i]
    ax.plot(tune_sample_arr[i,:curr_iter], gap_multiarr[i,:curr_iter], 
           label="T=%d" % (tune_explore_T_arr[i]),
           color=color_arr[i], linestyle=lss_arr[i])
    
ax.set(
    title="SPMD tuning noisy performance\n on a simple Battery problem",
    ylabel="Optimality gap",
    xlabel="Samples",
    xlim=(-1e5,4e6),
)
ax.legend()
```

```python
# true cost
i = np.argmax(alt_curr_iter_arr)
print("best performing tune_id = %d with T=%d" % (i, alt_tune_explore_T_arr[i]))

plt.style.use('ggplot')
_, ax = plt.subplots()

for i in range(n_tune):
    curr_iter = alt_curr_iter_arr[i]
    ax.plot(alt_tune_sample_arr[i,:curr_iter], alt_gap_multiarr[i,:curr_iter], 
           label="T=%d" % (alt_tune_explore_T_arr[i]),
           color=color_arr[i], linestyle=lss_arr[i])
    
ax.set(
    title="SPMD tuning performance\n on a simple Battery problem",
    ylabel="Optimality gap",
    xlabel="Samples",
    xlim=(-5e5,1.75e7),
)
ax.legend()
```
